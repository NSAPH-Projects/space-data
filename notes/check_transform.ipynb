```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_parquet("../trained_spaceenvs/pm25_simhosp_county_cont/data_full.parquet")
df_t = pd.read_parquet("../trained_spaceenvs/pm25_simhosp_county_cont/data_full_transformed.parquet")
```

```python
def plot_histograms(df, columns, bins=30):
    """
    Plots histograms for the specified columns in the given DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data.
        columns (list): List of column names to plot histograms for.
        bins (int): Number of bins for the histogram (default is 30).
    """
    for col in columns:
        if col in df.columns:
            plt.figure(figsize=(6, 4))
            plt.hist(df[col].dropna(), bins=bins, edgecolor='black', alpha=0.7)
            plt.title(f"Histogram of {col}")
            plt.xlabel(col)
            plt.ylabel("Frequency")
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.show()
        else:
            print(f"Warning: Column '{col}' not found in DataFrame.")
```

```python
plot_histograms(df_t, columns=df.columns[:18])
```
